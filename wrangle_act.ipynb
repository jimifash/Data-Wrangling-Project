3. Use the Tweepy library to query additional data via the Twitter API (tweet_json.txt)

```python
count=0
status=[]
fails_dict={}
start = timer()
for tweet_id in df1["tweet_id"]:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            s= api.get_status(tweet_id, tweet_mode='extended')
            status.append(s)
            print("Success")
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
print(fails_dict)
end = timer()
print(end - start)
print(fails_dict)
```

```python
list_=[]
for item in status:
    list_.append(item._json)
with open('tweet_json.txt','w') as file:
        file.write(json.dumps(list_, indent=True))
```

### Tidiness issues:

#### merging doggo,floofer,pupper and puppo columns into a single column

```python

#merging doggo,floofer,pupper and puppo columns into a single column
df1_clean['stage']= df1['doggo' ]+ ',' + df1['floofer']  + ',' + df1['pupper'] + ',' + df1['puppo']
df1_clean['stage']=df1_clean['stage'].str.split(',')


my_dict = {"new_stage":[]};
lists=['doggo','pupper','floofer','puppo']
for item in df1_clean['stage']:
    if lists[0] in item:
        my_dict["new_stage"].append(lists[0])
    elif lists[1] in item:
        my_dict["new_stage"].append(lists[1])
    elif lists[2] in item:
        my_dict["new_stage"].append(lists[2])
    elif lists[3] in item:
        my_dict["new_stage"].append(lists[3])
    else:
        my_dict["new_stage"].append(np.nan)

my_dict=pd.DataFrame(my_dict)
my_dict
```

```python
df1_clean=pd.concat([df1_clean,my_dict],axis=1)
```

```python
##  Dropping doggo,floofer,pupper, puppo and stage   columns
df1_clean.drop(['doggo','puppo','pupper','floofer','stage'],axis=1, inplace=True)
```

## Do people prefer storing dogs they like as favorite or retweeting

```python

total_retweet= df3.retweet_count.sum()
total_retweet
```

```python
total_favorite = df3.favorite_count.sum()
total_favorite
```

```python
locations = [1, 2]
heights = [total_retweet, total_favorite]
labels = ['total_retweet', 'total_favorite']
plt.bar(locations, heights, tick_label=labels)
plt.title("Relationship between total favorite and total retweet",fontsize=15)
plt.ylabel("frequency")
```

The rate at which users store the dog pic they like as favorite far exceeds the rate of retweet. This could be more people are more inclined to return to their favorite list to take another look.

### What stage of dog has the highest favorite count

```python
first.groupby('new_stage')['favorite_count'].mean().plot(kind='bar')
plt.title("Relation between Dog stage and favorite_count", fontsize=15)
plt.xlabel("Dog stage",fontsize=10)
plt.ylabel("mean favorite count",fontsize=10)
```

The graph shows that more people liked pics of dogs at the puppo stage.
